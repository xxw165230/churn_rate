```python
#Since, sklearn requires all inputs to be numeric,
#we should convert all our categorical variables into numeric by encoding the categories. 
#This can be done using the following code:

from sklearn.preprocessing import LabelEncoder

var_mode = df
le = LabelEncoder()
for i in var_mode:
    var_mode[i] = le.fit_transform(var_mode[i])
var_mode.dtypes 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import KFold   #For K-fold cross validation
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics
```

```python
#Generic function for making a classification model and accessing performance:

def classification_model(model, data, predictors, outcome):
#Fit the model:
    model.fit(data[predictors],data[outcome])
    
#Make predictions on training set:
    predictions = model.predict(data[predictors])

#Print accuracy
    accuracy = metrics.accuracy_score(predictions,data[outcome])
    print("Accuracy : %s" % "{0:.3%}".format(accuracy))

  
    
```

```python
outcome_var = 'Churn'
model = LogisticRegression()
predictor_var = ['gender','SeniorCitizen','Partner','Dependents','tenure','PhoneService','MultipleLines','InternetService']
classification_model(model, var_mode,predictor_var,outcome_var)
```

```python
outcome_var = 'Churn'
model = DecisionTreeClassifier()
predictor_var = ['gender','SeniorCitizen','Partner','Dependents','tenure','PhoneService','MultipleLines','InternetService']
classification_model(model, var_mode,predictor_var,outcome_var)
```

```python
outcome_var = 'Churn'
model = RandomForestClassifier(n_estimators=100)
predictor_var = ['gender','SeniorCitizen','Partner','Dependents','tenure','PhoneService','MultipleLines','InternetService']
classification_model(model, var_mode,predictor_var,outcome_var)
```
